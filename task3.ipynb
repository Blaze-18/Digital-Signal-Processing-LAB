 Show that the highest rate of oscillation in a discrete-time sinusoidal is obtained

```python
import numpy as np
import matplotlib.pyplot as plt
```

Steps to Show:
Discrete-time sinusoidal signal:
A discrete-time sinusoidal signal is of the form:

x[n]=Acos(ωn+ϕ)
where:
A is the amplitude,
ω is the angular frequency in radians,
n is the discrete time index.

Highest rate of oscillation: The rate of oscillation in a discrete-time signal corresponds to how many times the signal goes through one complete cycle per unit of time. The oscillation frequency increases as ω increases.

Key observation: When 
ω=π, the signal completes one cycle per 2 samples (i.e., the signal goes through two zero crossings per period), and the oscillation rate is maximized in discrete time.

```python


# Define parameters
A = 1          # Amplitude
n = np.arange(0, 50, 1)  # Discrete time index
omega_values = [np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]  # Different omega values

# Create a plot for different omega values
plt.figure(figsize=(12, 8))

for i, omega in enumerate(omega_values):
    x_n = A * np.cos(omega * n)  # Discrete-time sinusoidal signal
    plt.subplot(len(omega_values)+1, 2, i + 1)
    plt.plot(n, x_n, label=f'ω = {omega}')
    plt.title(f'Discrete-time sinusoidal signal (ω = {omega})')
    plt.xlabel('n (Discrete Time)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()
```
