Task Name: Generating elementary signals like Unit Step, Ramp, Exponential, Sine, and Cosine sequences.

```python
import numpy as np
import matplotlib.pyplot as plt
```

First we create the time axis and take evenly spread points on the time axis. For example -5s to 5s and 100 even points spread in this limit. All signals will be represented by this that is fro -5 to 5 seconds and they will have 1000 samples each

NOTE: DON'T USE LESS THAN 1000 Samples if the limit is from -5 to 5

```python
t = np.linspace(-5,6,1000) # Use 1000 points
```

Unit Step Function: Unit step function is a mathematical function that is 0 for all negative inputs and 1 for all positive inputs
u(t) = 0, for t < 0
u(t) = 1, for t > 0

Returns 1 when t ≥ t0, 0 otherwise
Default step occurs at t=0

```python
def unit_step(t, t0=0):
    return np.where(t>=0, 1.0, 0.0)

u = unit_step(t, 0)
```

A unit ramp signal is a continuous-time signal where the amplitude linearly increases with time, starting from zero. It's defined as r(t) = t for t >= 0 and r(t) = 0 for t < 0.

Returns a line with given slope starting at t0, 0 before t0
Default slope=1, starts at t=0

```python
def unit_ramp(t, slope=1, t0=0):
    return np.where(t >= t0, slope*(t-t0), 0.0)

r = unit_ramp(t, 1, 0)
```

An exponential signal is a signal that increases or decreases exponentially over time, described by the mathematical form s(t) = e^(αt). A positive value of 'α' indicates growth, while a negative 'α' indicates decay

```python
def exponential(t, a=1):
    return np.exp(a*t)
#growing
exp_pos = exponential(t, 0.5)
#decaying
exp_neg = exponential(t, -0.5)
```

Sine and Cosine Wave are trigonometric functions

sine(2*pi*f*t + theta)

cosine(2*pi*f*t + theta)

```python
def sine_wave(t, freq=1, phase=0):
    return np.sin(2*np.pi*freq*t + phase)

def cos_wave(t, freq=1, phase=0):
    return np.cos(2*np.pi*freq*t + phase)

sin = sine_wave(t, 0.5)
cos = cos_wave(t, 0.5)
```

Plotting All signals

```python
plt.figure(figsize=(15,10))

plt.subplot(3,2,1)
plt.plot(t, u)
plt.title('Unit Step Function')
plt.grid(True)

plt.subplot(3,2,2)
plt.plot(t, r)
plt.title('Unit Ramp Function')
plt.grid(True)

plt.subplot(3,2,3)
plt.plot(t, exp_pos, label='Growing (a=0.5)')
plt.plot(t, exp_neg, label='Decaying (a=-0.5)')
plt.title('Exponential Function')
plt.legend()
plt.grid(True)

plt.subplot(3,2,4)
plt.plot(t, sin)
plt.title('Sine Wave (0.5 Hz)')
plt.grid(True)

plt.subplot(3, 2, 5)
plt.plot(t, cos)
plt.title('Cosine Wave (0.5 Hz)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
