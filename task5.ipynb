 Consider the analog signal: xa(t)=3cos(200πt)+5sin(600πt)+10cos(1200πt).
Show the effect of sampling rate. 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#given signal = 3*cos(200*pi*t) + 5*sin(600*pi*t) + 10* cos(1200*pi*t)
f1=100
f2=300 
f3=600
def signal(t):
    return (3 * (np.cos(2*np.pi*f1*t)) ) + (5 *( np.sin(2*np.pi*f2*t) ) ) + 10 * (np.cos(2*np.pi*f3*t))
t_analog = np.linspace(0,0.01,1000)
```

```python
# showing the oriiginal signal
plt.figure(figsize=(10,4))
plt.title('Original Signal')
plt.plot(t_analog, signal(t_analog))
plt.grid(True)
plt.xlabel('time')
plt.ylabel('amplitude')
```

```python
# Showing diffrent sampliing effect
fs_values = [1500, 1200, 700]
plt.figure(figsize=(20,8))

for i, fs in enumerate(fs_values):
    ts = 1/fs
    t_sampled = np.arange(0, 0.01, ts)
    sampled_signal = signal(t_sampled)

    plt.subplot(len(fs_values)+1, 1, i+1)
    plt.plot(t_analog, signal(t_analog), 'grey')
    plt.stem(t_sampled,sampled_signal, basefmt=" ", linefmt='r-' )
    plt.title(f'Sampling at {fs} frequency')
    plt.subplots_adjust(hspace=0.5)
    plt.grid(True)




```
