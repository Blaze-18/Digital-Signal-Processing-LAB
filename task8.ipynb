# 6-Point Averaging Filter

This notebook demonstrates the use of a 6-point averaging filter on a composite signal. The original signal is made by combining two sine waves of different frequencies. Then a moving average filter is applied to smooth out the signal, reducing the effect of high-frequency components.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Input Signal

```python
sampling_rate = 100
t = np.linspace(0, 1, sampling_rate)

freq1 = 50  # High frequency
freq2 = 5   # Low frequency

sine_wave_x = np.sin(2 * np.pi * freq1 * t)
sine_wave_y = np.sin(0.5 * np.pi * freq2 * t)

input_wave = sine_wave_x + sine_wave_y
```

## Apply 6-Point Averaging Filter

```python
window_size = 6
filtered_wave = []

for i in range(len(input_wave) - window_size + 1):
    window = input_wave[i: i + window_size]
    average = np.sum(window) / window_size
    filtered_wave.append(average)

averaged_wave = np.array(filtered_wave)
t_new = t[:len(averaged_wave)]
```

## Plot the Original and Filtered Signal

```python
plt.figure(figsize=(10, 6))
plt.plot(t, input_wave, label='Input Signal')
plt.plot(t_new, averaged_wave, label='Filtered Signal (6-point average)')
plt.title('6-Point Averaging Filter')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
