 Given x(n)=[1,3,−2,4] y(n)=[2,3,−1,3] z(n)=[2,−1,4,−2]

Find the correlation between x(n) & y(n) and y(n) & z(n). ⟹ observe the
realization

```python
import numpy as np
import matplotlib.pyplot as plt

# Given signals
x = [1, 3, -2, 4]
y = [2, 3, -1, 3]
z = [2, -1, 4, -2]

# Manual correlation function
def manual_correlation(a, b):
    len_a = len(a)
    len_b = len(b)
    result_len = len_a + len_b - 1
    result = []

    # Pad a with zeros for safe indexing
    a_padded = [0] * (len_b - 1) + a + [0] * (len_b - 1)

    for lag in range(result_len):
        sum_val = 0
        for i in range(len_b):
            sum_val += a_padded[lag + i] * b[i]
        result.append(sum_val)

    return result

# Correlation between x(n) and y(n)
corr_xy = manual_correlation(x, y)
lags_xy = np.arange(-(len(y)-1), len(x))

# Correlation between y(n) and z(n)
corr_yz = manual_correlation(y, z)
lags_yz = np.arange(-(len(z)-1), len(y))

# Plot both
plt.figure(figsize=(12, 5))

# Plot x(n) and y(n)
plt.subplot(1, 2, 1)
plt.stem(lags_xy, corr_xy, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.title("Correlation between x(n) and y(n)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid(True)

# Plot y(n) and z(n)
plt.subplot(1, 2, 2)
plt.stem(lags_yz, corr_yz, basefmt=" ", linefmt='g-', markerfmt='go')
plt.title("Correlation between y(n) and z(n)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid(True)

plt.tight_layout()
plt.show()
```
