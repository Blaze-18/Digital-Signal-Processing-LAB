# Sampling and Aliasing Demonstration

This notebook demonstrates the effects of sampling a continuous-time signal. First, we plot the original continuous signal. Then we sample it at a frequency higher than the Nyquist rate, and finally, we sample it below the Nyquist rate to observe aliasing.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Original Continuous-Time Signal

```python
f = 5  # signal frequency in Hz
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * f * t)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Original Signal')
plt.title('Original Continuous-Time Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

## Sampling Above Nyquist Rate (No Aliasing)

```python
fs1 = 20  # Sampling frequency > 2*f
Ts1 = 1 / fs1
n1 = np.arange(0, 1, Ts1)
x_sampled1 = np.sin(2 * np.pi * f * n1)

plt.figure(figsize=(10, 4))
plt.plot(t, x, 'gray', alpha=0.5)
plt.stem(n1, x_sampled1)
plt.title('Sampling Above Nyquist Rate (No Aliasing)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

## Sampling Below Nyquist Rate (Aliasing)

```python
fs2 = 6  # Sampling frequency < 2*f
Ts2 = 1 / fs2
n2 = np.arange(0, 1, Ts2)
x_sampled2 = np.sin(2 * np.pi * f * n2)

plt.figure(figsize=(10, 4))
plt.plot(t, x, 'gray', alpha=0.5)
plt.stem(n2, x_sampled2)
plt.title('Sampling Below Nyquist Rate (Aliasing)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
