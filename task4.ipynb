```python
import numpy as np
import matplotlib.pyplot as plt

# Define the continuous-time signal
def x(t):
    return 3 * np.cos(100 * np.pi * t)

# Time range for continuous signal
t_cont = np.linspace(0, 0.05, 1000)
x_cont = x(t_cont)

# Sampling rates
fs_1 = 200  # Hz (above Nyquist rate)
fs_2 = 75   # Hz (below Nyquist rate)

# Sampling times
ts_1 = np.arange(0, 0.05, 1/fs_1)
ts_2 = np.arange(0, 0.05, 1/fs_2)

# Sampled signals
x_samp_1 = x(ts_1)
x_samp_2 = x(ts_2)

# Plotting
plt.figure(figsize=(14, 6))

# Plot for 200 Hz sampling
plt.subplot(1, 2, 1)
plt.plot(t_cont, x_cont, 'gray', label='Original Signal')
plt.stem(ts_1, x_samp_1, basefmt=" ", linefmt='b-', markerfmt='bo', label='Sampled @ 200Hz')
plt.title('Sampling at 200 Hz (Above Nyquist)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Plot for 75 Hz sampling
plt.subplot(1, 2, 2)
plt.plot(t_cont, x_cont, 'gray', label='Original Signal')
plt.stem(ts_2, x_samp_2, basefmt=" ", linefmt='r-', markerfmt='ro', label='Sampled @ 75Hz')
plt.title('Sampling at 75 Hz (Below Nyquist  Aliasing)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
