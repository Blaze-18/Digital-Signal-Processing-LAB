The impulse response of a discrete-time LTI system is h(n)={u(n)−u(n−5)}.
Determine the output of the system for the input x[n]=u(n), using the convolution
sum. 


```python

import numpy as np
import matplotlib.pyplot as plt


N = 20


def unit_step(n):
    return np.where(n >= 0, 1, 0)

n = np.arange(N)

# x[n] = u[n]
x = unit_step(n)

# h[n] = u[n] - u[n - 5]
h = unit_step(n) - unit_step(n - 5)

# Convolution using for loop
y = np.zeros(len(x) + len(h) - 1)
for i in range(len(x)):
    for j in range(len(h)):
        y[i + j] += x[i] * h[j]

# Output time axis
n_y = np.arange(len(y))

plt.figure(figsize=(12, 4))
plt.stem(n_y, y, basefmt=" ")
plt.title("Output y[n] using Convolution Sum")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)
plt.show()
```
